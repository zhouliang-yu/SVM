```python
import re
import numpy as np
from sklearn import datasets
from sklearn.multiclass import OutputCodeClassifier
from sklearn.svm import LinearSVC
import matplotlib.pyplot as plt
from sklearn import svm, datasets

class SVM(object):

    def __init__(self, training_dataset_, test_dataset_):
        self.training_dataset = training_dataset_
        self.test_dataset = test_dataset_
        self.classes = {}
        self.X_train = None
        self.Y_train = None
        self.X_test = None
        self.Y_test = None

        self.support_indecies = None
        self.train_errors = None
        self.val_errors = None
        self.train_loss = None
        self.val_loss = None
        
    def read_data(self):
        f = open(self.training_dataset, 'r')
        rows = list(re.split(' ', row) for row in re.split('\n', f.read())[:-1])
        names, self.Y_train = np.unique(list(row[-1] for row in rows), return_inverse=True)
        self.X_train = np.empty((0,4), float)
        f.close()
        for row in rows:
            self.X_train = np.append(self.X_train, np.array([np.array(row[:-1]).astype(float)]), axis = 0)
        f = open(self.test_dataset, 'r')
        f.close()
        
        f = open(self.test_dataset, 'r')
        rows = list(re.split(' ', row) for row in re.split('\n', f.read())[:-1])
        names, self.Y_test = np.unique(list(row[-1] for row in rows), return_inverse=True)
        self.X_test = np.empty((0,4), float)
        f.close()
        for row in rows:
            self.X_test = np.append(self.X_test, np.array([np.array(row[:-1]).astype(float)]), axis = 0)
        
    def SVM(self):
    
        #########################
        ## WRITE YOUR CODE HERE##
        #########################
        clf = svm.SVC(C=1e5, kernel = 'linear')
        clf.fit(self.X_train, self.Y_train) #training 
        train_score = clf.score(self.X_train, self.Y_train)

        num_error_train = 0
        predict_train = clf.predict(self.X_train)
        ## put the result into color plot

        for i in range(len(self.Y_train)):
            if (predict_train[i] != self.Y_train[i]):
                num_error_train += 1

        support_vector_indices_of_setosa = []
        support_vector_indices_of_versicolor = []
        support_vector_indices_of_virginica = []
        for i in clf.support_:
            if (self.Y_train[i] == 0):
                support_vector_indices_of_setosa.append(i)
            elif (self.Y_train[i] == 1):
                support_vector_indices_of_versicolor.append(i)
            elif (self.Y_train[i] == 2):
                support_vector_indices_of_virginica.append(i)
            else:
                break

        train_error = num_error_train / len(self.X_train)
        w = clf.coef_
        b = clf.intercept_
        w_of_setosa = w[0]
        w_of_versicolor = w[1]
        w_of_virginica = w[2]
        b_of_setosa = b[0]
        b_of_versicolor = b[1]
        b_of_virginica = b[2]

        pred_test = clf.predict(self.X_test)
        test_score = clf.score(self.X_test, self.Y_test)
        num_error_test = 0
        for i in range(len(self.Y_test)):
            if (pred_test[i] != self.Y_test[i]):
                num_error_test += 1

        test_error = num_error_test / len(self.X_test)
       
            
        return train_error, test_error, w_of_setosa, b_of_setosa, support_vector_indices_of_setosa, w_of_versicolor, b_of_versicolor, support_vector_indices_of_versicolor, w_of_virginica, b_of_virginica, support_vector_indices_of_virginica


    def SVM_slack(self, C):
    
        #########################
        ## WRITE YOUR CODE HERE##
        #########################    
        clf = svm.SVC(C = C, kernel = 'linear')
        clf.fit(self.X_train, self.Y_train) #training 
        train_score = clf.score(self.X_train, self.Y_train)
        
        num_error_train = 0
        predict_train = clf.predict(self.X_train)
        for i in range(len(self.Y_train)):
            if (predict_train[i] != self.Y_train[i]):
                num_error_train += 1

        support_vector_indices_of_setosa = []
        support_vector_indices_of_versicolor = []
        support_vector_indices_of_virginica = []
        for i in clf.support_:
            if (self.Y_train[i] == 0):
                support_vector_indices_of_setosa.append(i)
            elif (self.Y_train[i] == 1):
                support_vector_indices_of_versicolor.append(i)
            elif (self.Y_train[i] == 2):
                support_vector_indices_of_virginica.append(i)
            else:
                break

        train_error = num_error_train / len(self.X_train)
        w = clf.coef_
        b = clf.intercept_
        w_of_setosa = w[0]
        w_of_versicolor = w[1]
        w_of_virginica = w[2]
        b_of_setosa = b[0]
        b_of_versicolor = b[1]
        b_of_virginica = b[2]

        pred_test = clf.predict(self.X_test)
        test_score = clf.score(self.X_test, self.Y_test)
        num_error_test = 0
        for i in range(len(self.Y_test)):
            if (pred_test[i] != self.Y_test[i]):
                num_error_test += 1

        test_error = num_error_test / len(self.X_test)
        
        return train_error, test_error, w_of_setosa, b_of_setosa, support_vector_indices_of_setosa, w_of_versicolor, b_of_versicolor, support_vector_indices_of_versicolor, w_of_virginica, b_of_virginica, support_vector_indices_of_virginica

        
    
    def SVM_kernel_poly2(self, C):
    
        #########################
        ## WRITE YOUR CODE HERE##
        #########################    
        clf = svm.SVC(C = C, kernel = 'poly', degree = 2)
        clf.fit(self.X_train, self.Y_train) #training 
        train_score = clf.score(self.X_train, self.Y_train)
        
        num_error_train = 0
        predict_train = clf.predict(self.X_train)
        for i in range(len(self.Y_train)):
            if (predict_train[i] != self.Y_train[i]):
                num_error_train += 1

        support_vector_indices_of_setosa = []
        support_vector_indices_of_versicolor = []
        support_vector_indices_of_virginica = []
        for i in clf.support_:
            if (self.Y_train[i] == 0):
                support_vector_indices_of_setosa.append(i)
            elif (self.Y_train[i] == 1):
                support_vector_indices_of_versicolor.append(i)
            elif (self.Y_train[i] == 2):
                support_vector_indices_of_virginica.append(i)
            else:
                break

        train_error = num_error_train / len(self.X_train)
        b = clf.intercept_
        b_of_setosa = b[0]
        b_of_versicolor = b[1]
        b_of_virginica = b[2]

        pred_test = clf.predict(self.X_test)
        test_score = clf.score(self.X_test, self.Y_test)
        num_error_test = 0
        for i in range(len(self.Y_test)):
            if (pred_test[i] != self.Y_test[i]):
                num_error_test += 1

        test_error = num_error_test / len(self.X_test)
        
        return train_error, test_error, b_of_setosa, support_vector_indices_of_setosa, b_of_versicolor, support_vector_indices_of_versicolor,  b_of_virginica, support_vector_indices_of_virginica

        
    
    def SVM_kernel_poly3(self, C):
    
        #########################
        ## WRITE YOUR CODE HERE##
        #########################    
        
        clf = svm.SVC(C = C, kernel = 'poly', degree = 2)
        clf.fit(self.X_train, self.Y_train) #training 
        train_score = clf.score(self.X_train, self.Y_train)
        
        num_error_train = 0
        predict_train = clf.predict(self.X_train)
        for i in range(len(self.Y_train)):
            if (predict_train[i] != self.Y_train[i]):
                num_error_train += 1

        support_vector_indices_of_setosa = []
        support_vector_indices_of_versicolor = []
        support_vector_indices_of_virginica = []
        for i in clf.support_:
            if (self.Y_train[i] == 0):
                support_vector_indices_of_setosa.append(i)
            elif (self.Y_train[i] == 1):
                support_vector_indices_of_versicolor.append(i)
            elif (self.Y_train[i] == 2):
                support_vector_indices_of_virginica.append(i)
            else:
                break

        train_error = num_error_train / len(self.X_train)
        b = clf.intercept_
        b_of_setosa = b[0]
        b_of_versicolor = b[1]
        b_of_virginica = b[2]

        pred_test = clf.predict(self.X_test)
        test_score = clf.score(self.X_test, self.Y_test)
        num_error_test = 0
        for i in range(len(self.Y_test)):
            if (pred_test[i] != self.Y_test[i]):
                num_error_test += 1

        test_error = num_error_test / len(self.X_test)
        
        return train_error, test_error,  b_of_setosa, support_vector_indices_of_setosa, b_of_versicolor, support_vector_indices_of_versicolor, b_of_virginica, support_vector_indices_of_virginica

    
    def SVM_kernel_rbf(self, C):
    
        #########################
        ## WRITE YOUR CODE HERE##
        #########################    
        
        clf = svm.SVC(C = C, kernel = 'rbf')
        clf.fit(self.X_train, self.Y_train) #training 
        train_score = clf.score(self.X_train, self.Y_train)
        
        num_error_train = 0
        predict_train = clf.predict(self.X_train)
        for i in range(len(self.Y_train)):
            if (predict_train[i] != self.Y_train[i]):
                num_error_train += 1

        support_vector_indices_of_setosa = []
        support_vector_indices_of_versicolor = []
        support_vector_indices_of_virginica = []
        for i in clf.support_:
            if (self.Y_train[i] == 0):
                support_vector_indices_of_setosa.append(i)
            elif (self.Y_train[i] == 1):
                support_vector_indices_of_versicolor.append(i)
            elif (self.Y_train[i] == 2):
                support_vector_indices_of_virginica.append(i)
            else:
                break

        train_error = num_error_train / len(self.X_train)
        b = clf.intercept_

        b_of_setosa = b[0]
        b_of_versicolor = b[1]
        b_of_virginica = b[2]

        pred_test = clf.predict(self.X_test)
        test_score = clf.score(self.X_test, self.Y_test)
        num_error_test = 0
        for i in range(len(self.Y_test)):
            if (pred_test[i] != self.Y_test[i]):
                num_error_test += 1

        test_error = num_error_test / len(self.X_test)
        
        return train_error, test_error,  b_of_setosa, support_vector_indices_of_setosa, b_of_versicolor, support_vector_indices_of_versicolor, b_of_virginica, support_vector_indices_of_virginica

    
    def SVM_kernel_sigmoid(self, C):
    
        #########################
        ## WRITE YOUR CODE HERE##
        #########################    
        
        clf = svm.SVC(C = C, kernel = 'sigmoid')
        clf.fit(self.X_train, self.Y_train) #training 
        train_score = clf.score(self.X_train, self.Y_train)
        
        num_error_train = 0
        predict_train = clf.predict(self.X_train)
        for i in range(len(self.Y_train)):
            if (predict_train[i] != self.Y_train[i]):
                num_error_train += 1

        support_vector_indices_of_setosa = []
        support_vector_indices_of_versicolor = []
        support_vector_indices_of_virginica = []
        for i in clf.support_:
            if (self.Y_train[i] == 0):
                support_vector_indices_of_setosa.append(i)
            elif (self.Y_train[i] == 1):
                support_vector_indices_of_versicolor.append(i)
            elif (self.Y_train[i] == 2):
                support_vector_indices_of_virginica.append(i)
            else:
                break

        train_error = num_error_train / len(self.X_train)
        b = clf.intercept_
        b_of_setosa = b[0]
        b_of_versicolor = b[1]
        b_of_virginica = b[2]

        pred_test = clf.predict(self.X_test)
        test_score = clf.score(self.X_test, self.Y_test)
        num_error_test = 0
        for i in range(len(self.Y_test)):
            if (pred_test[i] != self.Y_test[i]):
                num_error_test += 1

        test_error = num_error_test / len(self.X_test)
        
        return train_error, test_error, b_of_setosa, support_vector_indices_of_setosa, b_of_versicolor, support_vector_indices_of_versicolor,  b_of_virginica, support_vector_indices_of_virginica

    
        
   




def output_file(out_name, result, func_type):

    if func_type == "svm" or func_type == "slack":

        training_error = result[0]
        testing_error = result[1]
        w_of_setosa = result[2]
        b_of_setosa = result[3]
        support_vector_indices_of_setosa = result[4]
        w_of_versicolor = result[5]
        b_of_versicolor = result[6]
        support_vector_indices_of_versicolor = result[7]
        w_of_virginica = result[8]
        b_of_virginica = result[9]
        support_vector_indices_of_virginica = result[10]
        f = open(out_name, "w+")
        f.write("training_error:%f\n" % training_error)
        f.write("testing_error:%f\n" % testing_error)
        f.write("w_of_setosa:")
        for i in w_of_setosa:
            f.write(str(i) + ", ")
            if (i == w_of_setosa[-1]):
                f.write("\n")

        f.write("b_of_setosa: %f\n" % b_of_setosa)
        f.write("support_vector_indices_of_setosa: ")
        for i in support_vector_indices_of_setosa:
            f.write(str(i) + ", ")
            if (i == support_vector_indices_of_setosa[-1]):
                f.write("\n")

        f.write("w_of_versicolor: ")
        for i in w_of_versicolor:
            f.write(str(i) + ", ")
            if (i == w_of_versicolor[-1]):
                f.write("\n")
        f.write("b_of_versicolor: %f\n" % b_of_versicolor)

        f.write("support_vector_indices_of_versicolor: ")
        for i in support_vector_indices_of_versicolor:
            f.write(str(i) + ", ")
            if (i == support_vector_indices_of_versicolor[-1]):
                f.write("\n")

        f.write("w_of_virginica: ")
        for i in w_of_virginica:
            f.write(str(i) + ", ")
            if (i == w_of_virginica[-1]):
                f.write("\n")

        f.write("b_of_virginica: %f\n" % b_of_virginica)
        f.write("support_vector_indices_of_virginica: ")

        for i in support_vector_indices_of_virginica:
            f.write(str(i) + ", ")
            if (i == support_vector_indices_of_virginica[-1]):
                f.write("\n")
        f.close()
    
    elif func_type == "nonlinear_kernel":

        training_error = result[0]
        testing_error = result[1]
        b_of_setosa = result[2]
        support_vector_indices_of_setosa = result[3]
        b_of_versicolor = result[4]
        support_vector_indices_of_versicolor = result[5]
        b_of_virginica = result[6]
        support_vector_indices_of_virginica = result[7]
        f = open(out_name, "w+")
        f.write("training_error:%f\n" % training_error)
        f.write("testing_error:%f\n" % testing_error)
        f.write("b_of_setosa: %f\n" % b_of_setosa)
        f.write("support_vector_indices_of_setosa: ")
        for i in support_vector_indices_of_setosa:
            f.write(str(i) + ", ")
            if (i == support_vector_indices_of_setosa[-1]):
                f.write("\n")
        f.write("b_of_versicolor: %f\n" % b_of_versicolor)

        f.write("support_vector_indices_of_versicolor: ")
        for i in support_vector_indices_of_versicolor:
            f.write(str(i) + ", ")
            if (i == support_vector_indices_of_versicolor[-1]):
                f.write("\n")

        f.write("b_of_virginica: %f\n" % b_of_virginica)
        f.write("support_vector_indices_of_virginica: ")

        for i in support_vector_indices_of_virginica:
            f.write(str(i) + ", ")
            if (i == support_vector_indices_of_virginica[-1]):
                f.write("\n")
        f.close()


    
```
